```python
#Question1:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, normalize
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
dataset = pd.read_csv('CC.csv')
dataset.head()
```

```python
dataset = dataset.drop("CUST_ID", axis=1)
```

```python
#Null values check
dataset.isnull().sum()
```

```python
#filling null values with 0s
dataset = dataset.fillna(0) 
```

```python
#applying PCA and took K(components)=2
pca = PCA(2)
pca.fit(dataset)
```

```python
x_pca = pca.fit_transform(dataset)
df2 = pd.DataFrame(data=x_pca)
```

```python
print(df2)
```

```python
#K-means after PCA
nclusters = 3 # this is the k in kmeans
km = KMeans(n_clusters=nclusters)
km.fit(df2)
```

```python
# predict the cluster for each data point
y_cluster_kmeans = km.predict(df2)
from sklearn import metrics
score = metrics.silhouette_score(df2, y_cluster_kmeans)
print(score)
```

```python
#Scaling main dataset
scaler = StandardScaler()
X_Scale = scaler.fit_transform(dataset)
```

```python
#PCA after feature scaling into original dataset
pca = PCA(2)#applying PCA and took K(components)=2
x_pca = pca.fit_transform(X_Scale)
```

```python
#Doing K-means after scaling+PCA
nclusters = 3           # this is the k in kmeans
km = KMeans(n_clusters=nclusters)
km.fit(x_pca)
```

```python
# predict the cluster for each data point
y_cluster_kmeans = km.predict(x_pca)
from sklearn import metrics
score = metrics.silhouette_score(x_pca, y_cluster_kmeans)
print(score)
```

```python
#Question2:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, normalize
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('pd_speech_features.csv')
dataset.head()
```

```python
scaler = preprocessing.StandardScaler()
scaler.fit(dataset)
dataset.head()
```

```python
X = dataset.drop('class',axis=1).values
y = dataset['class'].values
```

```python
pca2 = PCA(n_components=3)
principalComponents = pca2.fit_transform(X)

principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2','principal component 3'])

finalDf = pd.concat([principalDf, dataset[['class']]], axis = 1)
finalDf.head()
```

```python
X = finalDf.drop('class',axis=1).values
y = finalDf['class'].values
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3,random_state=0)
```

```python
# import SVC classifier
from sklearn.svm import SVC
# import metrics to compute accuracy
from sklearn.metrics import accuracy_score
# instantiate classifier with default hyperparameters
svc=SVC() 
# fit classifier to training set
svc.fit(X_train,y_train)
# make predictions on test set
y_pred=svc.predict(X_test)
```

```python
# compute and print accuracy score
accuracy = accuracy_score(y_test,y_pred)*100
print("accuracy of SVM:",accuracy)
```

```python
#Question3:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, normalize
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn import preprocessing
np.set_printoptions(precision=4)
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('Iris.csv')
dataset.head()
```

```python
#Encoding target column
from sklearn.preprocessing import LabelEncoder
class_le = LabelEncoder()
y = class_le.fit_transform(dataset["Species"].values)

x = dataset.iloc[:,range(0,4)].values
```

```python
#implementation of LDA with k = 2 
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis(n_components=2)
X_train_lda = lda.fit_transform(x, y)
```

```python
#Displaying result of LDA with k=2
data=pd.DataFrame(X_train_lda)
data['class']=y
data.columns=["LD1","LD2","class"]
data.head()
```

```python
#implementation of PCA with k = 2 
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)

principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])

finalDf = pd.concat([principalDf, data[['class']]], axis = 1)
finalDf.head()
```

```python
#difference between LDA and PCA
#LDA is useful when we have labels for data points and want to predict which label new points will have based on their 
#values. while if we don’t have labels then PCA will be useful.
#LDA is used for classification and PCA is used to get  low dimensional data. 
```
